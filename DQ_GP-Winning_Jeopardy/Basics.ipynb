Now Dataquest is recommending to create a function that normalizes words. Seems like a good idea in order to group and analyze them later.

```python
import string
#Defining function
def normalize_string(s):
    s_ = s
    s_ = str.lower(s_)
    s_ = s_.translate(str.maketrans('', '', string.punctuation))
    return s_

#A little test
normalize_string("Don't you EVER do. This. Again.")
```

```python
#Creating two Series with cleaned Question and Answer columns (I may still need the original columns, with punctuation)
clean_question = jeopardy['Question'].apply(normalize_string)
clean_answer = jeopardy['Answer'].apply(normalize_string)
```

```python
#Creating function to check if the answer is in the question
def answer_in_question(q, a):
    return a in q

#Testing it a bit
answer_in_question("dont you ever do something right","right")
```

Now checking if topics repeat themselves by checking how many times words with more than 6 characters repeat in different questions.

```python
word_frequency = {}
for r in clean_question:
    r_words = list(set(r.split(' ')))
    for w in r_words:
        if len(w) > 6:
            if w not in word_frequency:
                word_frequency[w] = 1
            else:
                word_frequency[w] += 1
```

```python
import operator
sorted_words = sorted(word_frequency.items(), key=operator.itemgetter(1), reverse=True)
sorted_words
```

From this list, we can get some topics that repeat themselves more frequently in questions. Country is by far the most repeated (2.4% of all our dictionary questions has it), with american, president and capital being very repeated too, which are mostly related to geopolitics questions.

```python
#Now checking how many questions doesn't have any words that repeat 20, 50 and 100 times or more (0.1% of all questions)
low_frequency_questions_20 = 0
low_frequency_questions_50 = 0
low_frequency_questions_100 = 0
for r in clean_question:
    r_words = list(set(r.split(' ')))
    max_w_freq = 0
    for w in r_words:
        if len(w) > 6:
            if word_frequency[w] > max_w_freq:
                max_w_freq = word_frequency[w]
    if max_w_freq < 20:
        low_frequency_questions_20 += 1
        low_frequency_questions_50 += 1
        low_frequency_questions_100 += 1
    elif max_w_freq < 50:
        low_frequency_questions_50 += 1
        low_frequency_questions_100 += 1
    elif max_w_freq < 100:
        low_frequency_questions_100 += 1
print("Lower than 20 words: " + str(low_frequency_questions_20/len(clean_question)))
print("Lower than 50 words: " + str(low_frequency_questions_50/len(clean_question)))
print("Lower than 100 words: " + str(low_frequency_questions_100/len(clean_question)))
```

So, there's 18% chance that a question will involve one of the terms of our dictionary with 100 or more repetitions. This chance grows to 37% when it comes to words appearing 50 or more times and to 59% when with words appearing 20 or more times, meaning that only 41% of all questions can be considered truly original (only with words that appear in less than 0.1% of all questions).

Now evaluating whether some of these words happen more in high-valued questions than lower ones.

```python
jeopardy_low = len(jeopardy[jeopardy['Value']<800])
jeopardy_low
```

```python
jeopardy_high = len(jeopardy[jeopardy['Value']>=800])
jeopardy_high
```

```python
jeopardy_high_expected = jeopardy_high/len(jeopardy)
jeopardy_high_expected
```

```python
#Selecting a list with only the most frequent words (>=100)
highest_words = sorted_words[:23]
highest_words
```

```python
#Seems like I'll have to work with qa_df first to check if a question is low or high valued
qa_df["High"] = (jeopardy["Value"] >= 800)
qa_df
```

```python
highest_words_observed = {}
for r in highest_words:
    w = r[0]
    count_low = 0
    count_high = 0
    for r2 in qa_df.iterrows():
        r2_columns = str(r2[1]).split(chr(10))
        r2_words = str(r2_columns[0]).split(' ')
        while '' in r2_words:
            r2_words.remove('')
        r2_value = str(r2_columns[3]).split(' ')
        while '' in r2_value:
            r2_value.remove('')
        #print(r2_value)
        if w in r2_words:
            if r2_value[1] == "False":
                count_low += 1
            else:
                count_high += 1
    highest_words_observed[w] = count_high/(count_high+count_low)
    print(highest_words_observed[w])
highest_words_observed
```

```python
chisq_highest_words = {}
observed_list = sorted(highest_words_observed.items(), key=operator.itemgetter(1), reverse=True)
for v in observed_list:
    chisq_highest_words[v[0]] = (float(v[1]) - jeopardy_high_expected)**2/jeopardy_high_expected
chisq_highest_words
```

Excluding the two "targetblank" values, which are some database mistake, one can conclude that "reports" are more frequent in high-valued questions, but it unfortunately doesn't tell much about the topic. "meaning" and "century" are more meaningful, since it refers to synonyms and history respectively.

```python
chances = []
for n in np.arange(0,1000):
    high = 0
    for i in np.arange(0,19999):
        if np.random.rand() <= jeopardy_high_expected:
            high += 1
    chances.append(high/19999)
chance_highest_words = {}
for v in observed_list:
    chance = 0
    for c in chances:
        if float(v[1]) < c:
            chance += 1
    chance_highest_words[v[0]] = chance/10
chance_highest_words
```

The list above shows the probability of the frequency of these words happening mostly in high question happening by chance, meaning that, for all words with 0%, there is a preference for them happening in high-valued questions over low-valued ones. Since, even for words that clearly have a higher frequency for high over low-valued questions, the chance of it being asked in a low-valued question is around 50% (like "century"), I wouldn't go with this data to recommend someone to study more this topic, it was good to know as a statistics point of view, but not very useful for the contester
